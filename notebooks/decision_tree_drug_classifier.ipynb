# 🩺 Project Overview: Drug Prescription Prediction using Decision Trees

<b>In this project, we aim to build a **multiclass classification model** using a Decision Tree algorithm to predict the most suitable drug for patients suffering from the same illness. The dataset includes patient information such as **Age, Sex, Blood Pressure, and Cholesterol levels**, and the target variable is the drug prescribed (**Drug A, B, C, X, or Y**).<b/>
<br/>
<b>The goal is to train a model that can accurately **recommend a drug** for new patients based on their medical attributes.<b/>


```python
#importing the required libraries
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
path= 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/drug200.csv'
my_data = pd.read_csv(path)
my_data
```

<b>Data Analysis and pre-processing<b/>

```python
my_data.info()
```

```python
#4 out of the 6 features of this dataset are categorical
#convert into numerical to be used for modeling
label_encoder=LabelEncoder()
my_data["Sex"]=label_encoder.fit_transform(my_data["Sex"])
my_data["BP"]=label_encoder.fit_transform(my_data["BP"])
my_data['Cholesterol'] = label_encoder.fit_transform(my_data['Cholesterol']) 
my_data
```

```python
# We are working with a dataset that contains 5 input features and 1 target variable.
# After applying Label Encoding to categorical features, we use the following mappings:

# Encoding for 'Sex':
#   F → 0
#   M → 1

# Encoding for 'BP' (Blood Pressure):
#   LOW → 0
#   NORMAL → 1
#   HIGH → 2

# Encoding for 'Cholesterol':
#   NORMAL → 0
#   HIGH → 1

# These encodings are necessary because machine learning models like Decision Trees
# require numerical input rather than categorical text labels.
```

```python
my_data.isnull().sum()
```

```python
#Map drug labels to numerical values for correlation analysis
# Custom mapping of drug names to integers:
custom_map = {
    'drugA': 0,
    'drugB': 1,
    'drugC': 2,
    'drugX': 3,
    'drugY': 4
}
# Create a new column 'Drug_num' by applying the mapping to the 'Drug' column
my_data['Drug_num'] = my_data['Drug'].map(custom_map)
my_data
```

```python
# Drug_num       1.000000      ← This is the target itself (correlation with itself is always 1.0)
# Na_to_K        0.589120      ← Strong positive correlation
# BP             0.372868      ← Moderate positive correlation
# Cholesterol    0.055629      ← Very weak positive correlation
# Age           -0.004828      ← Almost no correlation
# Sex           -0.098573      ← Very weak negative correlation
my_data.drop('Drug',axis=1).corr()['Drug_num']
#This shows that the drug recommendation is mostly correlated with the Na_to_K and BP features.
```

```python
category_counts = my_data['Drug'].value_counts()
category_counts
```

```python
# Plot the count plot
plt.bar(category_counts.index, category_counts.values, color='blue')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.title('Category Distribution')
plt.xticks(rotation=45)  # Rotate labels for better readability if needed
plt.show()
```

<b>Modeling<b/>

```python
# Separate the target variable and features
y = my_data['Drug']  # Target: actual drug prescribed (categorical)
X = my_data.drop(['Drug', 'Drug_num'], axis=1)  # Features only (exclude target and encoded target)
```

```python
# Split data into training and testing sets (70% train, 30% test)
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=32)
```

```python
# Create Decision Tree classifier object using entropy as the splitting criterion
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth=4)
```

```python
# Train the classifier on the training set
drugTree.fit(X_trainset, y_trainset)
```

<b>Evaluation<b/>

```python
# Generate predictions on the test set
tree_predictions = drugTree.predict(X_testset)

# Import accuracy metric
from sklearn import metrics

# Calculate and print accuracy of the model
accuracy = metrics.accuracy_score(y_testset, tree_predictions)
print("Decision Tree's Accuracy: {:.2f}%".format(accuracy * 100))

# This tells us how many of the test samples were correctly classified by the model
```

<b>Visualize the tree<b/>

```python
# Visualize the trained decision tree
plt.figure(figsize=(12,8))  # Set the size of the plot for better readability
plot_tree(drugTree, filled=True, rounded=True, 
          feature_names=X.columns,  # Show feature names on the nodes
          class_names=drugTree.classes_)  # Show the target class names on the leaves
plt.title("Decision Tree for Drug Prescription")
plt.show()
```

```python
# Train decision tree with reduced depth
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth=3)
drugTree.fit(X_trainset, y_trainset)

# Predictions and accuracy
tree_predictions = drugTree.predict(X_testset)
print("Decision Tree's Accuracy: ", metrics.accuracy_score(y_testset, tree_predictions))
```
